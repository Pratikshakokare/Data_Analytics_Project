### Data Visualization

```python
# plot isnull heatmap
sns.heatmap(covid_vacc.isnull())
plt.show()
```

```python
# in which state maximum number Registered cases
Max=covid_vacc.groupby('State')['Total_Registered_Cases'].sum().sort_values(ascending=False).head(10)
Max
```

```python
# seaborn bar plot 
# In  which State Maximum  number of Cases Registered
sns.set_style('whitegrid')
sns.set_context('paper')
sns.barplot(x=Max.index, y=Max,palette="rocket")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# in which state maximum number vaccination  conducted
vaccine=covid_vacc.groupby('State')['Total_Vaccinated'].sum().sort_values(ascending=False).head(10)
vaccine
```

```python
# In  which State Maximum  number of vaccination conducted
colors=['#ec7063' if (x < max(vaccine)) else '#C70039' for x in vaccine]
sns.set_style('whitegrid')
sns.set_context('paper')
sns.barplot(x=vaccine.index, y=vaccine,palette=colors)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Display the Count of all State
covid_vacc['State'].value_counts()
```

```python
# Plot the barplot of State Vs Count
State =covid_vacc.State.value_counts()
sns.set_style("darkgrid")
plt.figure(figsize=(10,10))
sns.barplot(y=State.index, x=State.values)
plt.show()
```

```python
# line Plot of Total_Vaccination Vs Total_Doses_Administered
plt.plot(covid_vacc.Total_Registered_Cases,covid_vacc. Total_Doses_Administered )
plt.plot(covid_vacc.Total_Registered_Cases,covid_vacc. Total_Vaccinated )
plt.title('Total Doses Administered over Vaccinated',color='red',fontsize=20,fontname='Times New Roman')
plt.xlabel('Total Vaccinated', color='orange',fontsize=20,fontname='Brush Script MT')
plt.ylabel('Total Doses Administered',color='orange',fontsize=20,fontname='Brush Script MT')
plt.legend(['Administered','Vaccinated'],loc=2)
plt.show()
```
